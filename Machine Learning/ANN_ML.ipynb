```python
import pandas as pd
```

```python
annual_us_gas_data = pd.read_csv('Resources/Datasets/cleaned_data/Annual_us_gas_data.csv')
consumer_factors = pd.read_csv('Resources/Datasets/cleaned_data/consumer_factors.csv')
more_factors = pd.read_csv('Resources/Datasets/cleaned_data/4_more_factors.csv')
```

```python
# rename columns
annual_us_gas_data = annual_us_gas_data.rename(columns = {
    'avg_production(Mmcf)':'production(Mmcf)', 
    'avg_consumption(Mmcf)': 'consumption(Mmcf)',
       'avg_vol(Mmcf)': 'volumn(Mmcf)', 
    'avg_total_import(Mmcf)': 'total_import(Mmcf)', 
    'avg_toal_export(Mmcf)': 'total_export',
       'avg_import_price($/Mmcf)': 'import_price($/Mmcf)', 
    'avg_export_price($/Mmcf)':'export_price($/Mmcf)',
       'avg_all_grades_p($/Gallon)':'all_grades_price($/Gallon)', 'avg_reg_p($/Gallon)':'reg_price($/Gallon)',
       'avg_midg_p($/Gallon)':'midg_price($/Gallon)', 'avg_pre_p($/Gallon)':'prem_price($/Gallon)',
    'avg_dis_p($/Gallon)':'dis_price($/Gallon)'
})

annual_us_gas_data
```

```python
## Merge data
ANN_data = annual_us_gas_data.merge(consumer_factors.merge(more_factors, on='Year'), on='Year')
ANN_data
```

```python
ANN_data = ANN_data.astype({'all_conusmers(Mmcf)':float,
                            'commercial_consumers(Mmcf)':float,
                            'electric_consumers(Mmcf)':float,
                           'industrial_consumers(Mmcf)':float, 
                            'residential_consumers(Mmcf)':float,
                           'vehicle_consumers(Mmcf)':float,
                            'working_population':float
                           })
```

# ANN model to analysis

```python
import matplotlib as plt
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import sklearn.metrics as skm
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#initial features and target
X = ANN_data.drop(columns = 'all_grades_price($/Gallon)')
y = ANN_data['all_grades_price($/Gallon)']

#split training 
X_train, X_test,y_train, y_test = train_test_split(X, y, random_state =42)
```

```python
X_train.shape
```

```python
#scale data
X_scaler = StandardScaler().fit(X_train)
X_train_s = X_scaler.transform(X_train)
X_test_s = X_scaler.transform(X_test)
```

```python
X_train_s.shape
```

```python
#create model
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(units = 46, activation = 'relu', input_dim=22))
nn_model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
nn_model.summary()
```

```python
nn_model.weights
```

```python
#compile the model and custom metrics
nn_model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mae','accuracy'])
```

```python
#train the model and fit
fit_model = nn_model.fit(X_train_s, y_train, epochs=50)
```

```python
#evaluate the data
model_loss_mse, model_mae = nn_model.evaluate(X_test_s, y_test, verbose=2)

print(f'MeanSquaredError: {model_loss_mse:.4f} , MeanAbsoluteError: {model_mae:.4f} .')
```
