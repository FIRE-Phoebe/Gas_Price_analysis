# Work on "Gas Price by Type" 

```python
import pandas as pd
```

```python
# Read Gas Price by Type
gas_price_by_type_df = pd.read_csv('Resources/gas_prices_by_type.csv')
gas_price_by_type_df.head(10)
```

```python
# convert our Date column to the right format -> datetime64
gas_price_by_type_df["Date"] = pd.to_datetime(gas_price_by_type_df["Date"])
gas_price_by_type_df.head()
```

```python
# change our Date column to Month using to_period
per = gas_price_by_type_df.Date.dt.to_period("M")
per
```

```python
# now groupby "per" and sum by month
g = gas_price_by_type_df.groupby(per)
gas_price_by_type_by_month_df = g.sum()
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df.reset_index(inplace=True)
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df= gas_price_by_type_by_month_df.rename(columns={"Date":"Year-Month"})
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df['Year'] = gas_price_by_type_by_month_df['Year-Month'].dt.year
gas_price_by_type_by_month_df['Month'] = gas_price_by_type_by_month_df['Year-Month'].dt.month
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df = gas_price_by_type_by_month_df.drop(columns = "Year-Month")
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df = gas_price_by_type_by_month_df[["Year","Month","A1","A2","A3","R1","R2","R3","M1","M2", "M3","P1","P2", "P3","D1"]]
gas_price_by_type_by_month_df
```

# California_Regular_All_Formulations_Retail_Gasoline_Prices

```python
import pandas as pd
```

```python
cali_all_form_retail_gas_prices_df = pd.read_csv('Resources/California_Regular_All_Formulations_Retail_Gasoline_Prices.csv',skiprows=4)
cali_all_form_retail_gas_prices_df.head(10)
```

```python
cali_all_form_retail_gas_prices_df["Date"] = pd.to_datetime(cali_all_form_retail_gas_prices_df["Month"])
cali_all_form_retail_gas_prices_df.head()
```

```python
cali_all_form_retail_gas_prices_df = pd.DataFrame(data=cali_all_form_retail_gas_prices_df, columns=["Date",
            "California Regular All Formulations Retail Gasoline Prices Dollars per Gallon"])
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df= cali_all_form_retail_gas_prices_df.rename(columns={"Date":"Year-Month",
        "California Regular All Formulations Retail Gasoline Prices Dollars per Gallon":"Price per Gallon"})
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df['Year'] = cali_all_form_retail_gas_prices_df['Year-Month'].dt.year
cali_all_form_retail_gas_prices_df['Month'] = cali_all_form_retail_gas_prices_df['Year-Month'].dt.month
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df.columns
```

```python
cali_all_form_retail_gas_prices_df= cali_all_form_retail_gas_prices_df.drop(["Year-Month"],axis=1)
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df= pd.DataFrame(data =cali_all_form_retail_gas_prices_df,
                                        columns={'Year':'Year',
                                                 'Month':'Month',
                                                'Price per Gallon': 'Price per Gallon'})
cali_all_form_retail_gas_prices_df
```
