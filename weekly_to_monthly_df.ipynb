# Work on "Gas Price by Type" 

```python
import pandas as pd
```

```python
# Read Gas Price by Type
gas_price_by_type_df = pd.read_csv('Resources/gas_prices_by_type.csv')
gas_price_by_type_df.head(10)
```

```python
# convert our Date column to the right format -> datetime64
gas_price_by_type_df["Date"] = pd.to_datetime(gas_price_by_type_df["Date"])
gas_price_by_type_df.head()
```

```python
# change our Date column to Month using to_period
per = gas_price_by_type_df.Date.dt.to_period("M")
per
```

```python
# now groupby "per" and sum by month
g = gas_price_by_type_df.groupby(per)
gas_price_by_type_by_month_df = g.sum()
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df.reset_index(inplace=True)
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df= gas_price_by_type_by_month_df.rename(columns={"Date":"Year-Month"})
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df['Year'] = gas_price_by_type_by_month_df['Year-Month'].dt.year
gas_price_by_type_by_month_df['Month'] = gas_price_by_type_by_month_df['Year-Month'].dt.month
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df = gas_price_by_type_by_month_df.drop(columns = "Year-Month")
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df = gas_price_by_type_by_month_df[["Year","Month","A1","A2","A3","R1","R2","R3","M1","M2", "M3","P1","P2", "P3","D1"]]
gas_price_by_type_by_month_df
```

# Work on "Weekly California Regular All Formulations Retail Gasoline Prices"

```python
cali_all_form_retail_gas_price_df = pd.read_csv('Resources/Weekly_California_Regular_All_Formulations_Retail_Gasoline_Prices.csv')
cali_all_form_retail_gas_price_df.head(10)
```

```python
cali_all_form_retail_gas_price_df=cali_all_form_retail_gas_price_df.drop(labels=[0,1,2,3], axis=0, inplace=False)
cali_all_form_retail_gas_price_df
```

```python
cali_all_form_retail_gas_price_df = cali_all_form_retail_gas_price_df.rename(columns=
            {"Weekly California Regular All Formulations Retail Gasoline Prices":"Date", "Unnamed: 1":"Price_per_Gallon"})
cali_all_form_retail_gas_price_df
```

```python
# convert our Date column to the right format -> datetime64
cali_all_form_retail_gas_price_df["Date"] = pd.to_datetime(cali_all_form_retail_gas_price_df["Date"])
cali_all_form_retail_gas_price_df.head()
```

```python
cali_all_form_retail_gas_price_df = cali_all_form_retail_gas_price_df.reset_index()
cali_all_form_retail_gas_price_df
```

```python
cali_all_form_retail_gas_price_df = cali_all_form_retail_gas_price_df.drop(columns=["index"])
cali_all_form_retail_gas_price_df
```

```python
cali_all_form_retail_gas_price_df["Price_per_Gallon"]=cali_all_form_retail_gas_price_df["Price_per_Gallon"].astype(float, errors="raise")
cali_all_form_retail_gas_price_df
```

```python
cali_all_form_retail_gas_price_df.dtypes
```

```python
per = cali_all_form_retail_gas_price_df.Date.dt.to_period("M")
per
```

```python
g = cali_all_form_retail_gas_price_df.groupby(per)
g
```

```python
cali_all_form_retail_gas_price_per_month = g.mean()
cali_all_form_retail_gas_price_per_month
```

```python
cali_all_form_retail_gas_price_per_month = cali_all_form_retail_gas_price_per_month.round(2)
cali_all_form_retail_gas_price_per_month
```

```python
cali_all_form_retail_gas_price_per_month.columns
```

```python
cali_all_form_retail_gas_price_per_month = cali_all_form_retail_gas_price_per_month.reset_index(inplace=True)
cali_all_form_retail_gas_price_per_month
```

```python
# Export data to new csv file 
cali_all_form_retail_gas_price_per_month.to_csv('Resources/cali_all_form_retail_gas_price_monthly.csv')
```
