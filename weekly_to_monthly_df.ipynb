# Work on "Gas Price by Type" 

```python
import pandas as pd
```

```python
# Read Gas Price by Type
gas_price_by_type_df = pd.read_csv('Resources/gas_prices_by_type.csv')
gas_price_by_type_df.head(10)
```

```python
# convert our Date column to the right format -> datetime64
gas_price_by_type_df["Date"] = pd.to_datetime(gas_price_by_type_df["Date"])
gas_price_by_type_df.head()
```

```python
# change our Date column to Month using to_period
per = gas_price_by_type_df.Date.dt.to_period("M")
per
```

```python
# now groupby "per" and sum by month
g = gas_price_by_type_df.groupby(per)
gas_price_by_type_by_month_df = g.sum()
gas_price_by_type_by_month_df
```

```python

gas_price_by_type_by_month_df.reset_index(inplace=True)
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df= gas_price_by_type_by_month_df.rename(columns={"Date":"Year-Month"})
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df['Year'] = gas_price_by_type_by_month_df['Year-Month'].dt.year
gas_price_by_type_by_month_df['Month'] = gas_price_by_type_by_month_df['Year-Month'].dt.month
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df = gas_price_by_type_by_month_df.drop(columns = "Year-Month")
gas_price_by_type_by_month_df
```

```python
gas_price_by_type_by_month_df = gas_price_by_type_by_month_df[["Year","Month","A1","A2","A3","R1","R2","R3","M1","M2", "M3","P1","P2", "P3","D1"]]
gas_price_by_type_by_month_df
```

# Work on Price of U.S. Natural Gas Exports

# Work on Price of U.S. Natural Gas Imports

```python
imports_price_df = pd.read_csv('Resources/Price_of_U.S._Natural_Gas_Imports.csv',skiprows=4)
imports_price_df.head(10)
```

# Merge Exports_df and imports_df

```python
us_price_export_import = imports_price_df.merge(exports_price_df, how='left', on = 'Month')
us_price_export_import
```

```python
us_price_export_import["Date"] = pd.to_datetime(us_price_export_import["Month"])
us_price_export_import.head()
```

```python
us_price_export_import= us_price_export_import.drop(["Month"],axis=1)
us_price_export_import
```

```python
us_price_export_import.columns
```

```python
us_price_export_import = pd.DataFrame(data=us_price_export_import, columns={'Date',
            "Price of U.S. Natural Gas Imports  Dollars per Thousand Cubic Feet",
            "Price of U.S. Natural Gas Exports  Dollars per Thousand Cubic Feet"})
us_price_export_import
```

```python
us_price_export_import=us_price_export_import.rename(columns={"Date":"Year-Month",
        "Price of U.S. Natural Gas Imports  Dollars per Thousand Cubic Feet":"Imports_Price_per_Gallon",
        "Price of U.S. Natural Gas Exports  Dollars per Thousand Cubic Feet":"Exports_Price_per_Gallon"})
us_price_export_import
```

```python
us_price_export_import['Year'] = us_price_export_import['Year-Month'].dt.year
us_price_export_import['Month'] = us_price_export_import['Year-Month'].dt.month
us_price_export_import
```

```python
us_price_export_import.columns
```

```python
us_price_export_import= us_price_export_import.drop(["Year-Month"],axis=1)
us_price_export_import
```

```python
us_price_export_import= pd.DataFrame(data =us_price_export_import,
                                        columns={'Year':'Year',
                                                 'Month':'Month',
                                                'Exports_Price_per_Gallon': 'Exports_Price_per_Gallon',
                                                'Imports_Price_per_Gallon': 'Imports_Price_per_Gallon'})
us_price_export_import
```
