```python
import pandas as pd
```

## Clean U.S Consumption data

```python
#read data gas consumption
gas_consumption_df = pd.read_csv('Gas consumption by country.csv')
gas_consumption_df.head()
```

```python
gas_consumption_df.columns
```

```python
gas_consumption_df.count()
```

```python
gas_consumption_df.isna().sum()
```

```python
#drop nas
consumption_df = gas_consumption_df.drop(columns='Code')
consumption_df.head()
```

```python
consumption_df.Entity.unique()
```

```python
consumption_df.dtypes
```

```python
#filter the gas price shows only U.S and Year from 2020-current
us_consumption_2000 = consumption_df[(consumption_df.Entity == 'United States') & (consumption_df.Year >= int('2000'))]
us_consumption_2000
```

```python
#rename consumption columns
us_consumption_2000= us_consumption_2000.rename(columns={'Gas Consumption - TWh': 'Gas_Consumption(TWh)'})

#set Year as index and drop entity
us_consumption_2000 = us_consumption_2000.set_index('Year')
us_gas_consumption_df = us_consumption_2000.drop(columns = 'Entity')
```

```python
#check the new df
us_gas_consumption_df 
```

## Clean U.S Gas Procudtion data

```python
# import data production
production_df = pd.read_csv('Gas production by country.csv')
production_df.head()
```

```python
#rename columns and set index
production_df = production_df.rename(columns={'Gas production (TWh)' : 'Gas_production(TWh)'})
production_df = production_df.set_index('Year')
production_df.head()
```

```python
production_df.count()
```

```python
#drop Code
production_df = production_df.drop(columns='Code')
production_df.head()
```

```python
#filter data for U.S.since 2000
us_production_df = production_df[(production_df.Entity == 'United States') & (production_df.index >= int('2000'))]
us_production_df
```

```python
#drop entity column
us_gas_production_df = us_production_df.drop(columns='Entity')
us_gas_production_df
```

## Create new dataframe has consumption and production data

```python
## Merge Dataframe for US gas consumption and production
new_df = us_gas_consumption_df.merge(us_gas_production_df, how = 'left', on = 'Year')
new_df
```
