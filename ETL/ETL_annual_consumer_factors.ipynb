```python
import pandas as pd
```

```python
all_consumers = pd.read_csv('../Resources/Datasets/Consumer_factor_data/All_Consumers_in_the_U.S..csv', skiprows=4)
all_consumers.columns
```

```python
#rename columns
all_consumers = all_consumers.rename(columns={
    'Natural Gas Delivered to Consumers in the U.S.  Million Cubic Feet': 'all_conusmers(Mmcf)' 
})

all_consumers
```

```python
#filter data from 2001-recent
all_consumers = all_consumers[(all_consumers.Year >= int(2001))]
```

```python
all_consumers
```

```python
commercial_consumers = pd.read_csv('../Resources/Datasets/Consumer_factor_data/Commercial_Consumers.csv', skiprows=4)
commercial_consumers.columns
```

```python
#rename columns
commercial_consumers = commercial_consumers.rename(columns = {
    'Natural Gas Deliveries to Commercial Consumers (Including Vehicle Fuel through 1996) in the U.S.  Million Cubic Feet' : 'commercial_consumers(Mmcf)'
})

# filter data
commercial_consumers =commercial_consumers[(commercial_consumers.Year>= int(2001))]
commercial_consumers
```

```python
electric_consumers = pd.read_csv('../Resources/Datasets/Consumer_factor_data/Electric_Power_Consumers.csv',skiprows=4)
electric_consumers.columns
```

```python
#rename columns
electric_consumers = electric_consumers.rename(columns = {
   'U.S. Natural Gas Deliveries to Electric Power Consumers  Million Cubic Feet': 'electric_consumers(Mmcf)' 
})

#filter data
electric_consumers = electric_consumers[(electric_consumers.Year >= int(2001))]
electric_consumers
```

```python
industrial_consumers = pd.read_csv('../Resources/Datasets/Consumer_factor_data/Industrial_Consumption.csv',skiprows=4)
industrial_consumers.columns
```

```python
#rename columns
industrial_consumers = industrial_consumers.rename(columns = {
    'U.S. Natural Gas Industrial Consumption  Million Cubic Feet':'industrial_consumers(Mmcf)'
})

#filter data
industrial_consumers = industrial_consumers[(industrial_consumers.Year>= int(2001))]
industrial_consumers
```

```python
residential_consumers = pd.read_csv('../Resources/Datasets/Consumer_factor_data/Residential_Consumption.csv',skiprows=4)
residential_consumers.columns
```

```python
#rename columns
residential_consumers = residential_consumers.rename(columns = {
    'U.S. Natural Gas Residential Consumption  Million Cubic Feet':'residential_consumers(Mmcf)'
})

#filter data
residential_consumers = residential_consumers[(residential_consumers.Year >= int(2001))]
residential_consumers
```

```python
vehicle_consumers = pd.read_csv('../Resources/Datasets/Consumer_factor_data/Vehicle_Fuel_Consumption.csv',skiprows=4)
vehicle_consumers.columns
```

```python
#rename columns
vehicle_consumers = vehicle_consumers.rename(columns = {
    'U.S. Natural Gas Vehicle Fuel Consumption  Million Cubic Feet': 'vehicle_consumers(Mmcf)'
})

#filter data
vehicle_consumers = vehicle_consumers[(vehicle_consumers.Year >= int(2001))]
vehicle_consumers
```

```python
# Merge all data to dataframe
consumer_factors = all_consumers.merge(commercial_consumers.merge(electric_consumers.merge(industrial_consumers.merge(residential_consumers.merge(vehicle_consumers, on='Year'), on='Year'), on='Year'), on='Year'), on='Year')
consumer_factors
```

```python
##save to consumer factors
consumer_factors.to_csv('../Resources/Datasets/cleaned_data/consumer_factors.csv', index=False)
```
