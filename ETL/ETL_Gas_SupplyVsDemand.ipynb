```python
import pandas as pd
```

```python
market_production = pd.read_csv('U.S._Natural_Gas_Marketed_Production.csv', skiprows=4)
market_production.head()
```

```python
market_production.columns
```

```python
#rename columns
market_production = market_production.rename(columns= {
                    'Month': 'Period', 'U.S. Natural Gas Marketed Production  Million Cubic Feet': 'Gas_Procution(Mmcf)'
})

market_production.head()
```

```python
consumption_df = pd.read_csv('U.S._Natural_Gas_Total_Consumption.csv', skiprows=4)
consumption_df.head()
```

```python
consumption_df.columns
```

```python
#rename columns
consumption_df = consumption_df.rename(columns={'Month': 'Period', 'U.S. Natural Gas Total Consumption  Million Cubic Feet':'Gas_Consumption(Mmcf)'
    })

consumption_df.head()
```

```python
## Merge to df
production_consumption = market_production.merge(consumption_df, how='left', on='Period')
production_consumption.head()
```

```python
#extract year and month
production_consumption['Period'] = pd.to_datetime(production_consumption['Period'])
production_consumption['Year'] = production_consumption['Period'].dt.year
production_consumption['Month'] = production_consumption['Period'].dt.month
production_consumption.head()
```

```python
production_consumption.isnull().sum()
```

```python
#reorder columns
supply_demand_df = pd.DataFrame(data =production_consumption,
                                        columns={'Year':'Year',
                                                 'Month':'Month',
                                                'Gas_Procution(Mmcf)': 'Gas_Procution(Mmcf)',
                                               'Gas_Consumption(Mmcf)':'Gas_Consumption(Mmcf)'})

supply_demand_df
```

```python
#drop nas
supply_demand_df = supply_demand_df.dropna()
supply_demand_df
```

## Save as csv file

```python
supply_demand_df.to_csv('supply_demand.csv')
```
