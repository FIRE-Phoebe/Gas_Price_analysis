# California_Regular_All_Formulations_Retail_Gasoline_Prices

```python
import pandas as pd
```

```python
cali_all_form_retail_gas_prices_df = pd.read_csv('Resources/California_Regular_All_Formulations_Retail_Gasoline_Prices.csv',skiprows=4)
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df["Date"] = pd.to_datetime(cali_all_form_retail_gas_prices_df["Month"])
cali_all_form_retail_gas_prices_df.head()
```

```python
cali_all_form_retail_gas_prices_df = pd.DataFrame(data=cali_all_form_retail_gas_prices_df, columns=["Date",
            "California Regular All Formulations Retail Gasoline Prices Dollars per Gallon"])
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df= cali_all_form_retail_gas_prices_df.rename(columns={"Date":"Year-Month",
        "California Regular All Formulations Retail Gasoline Prices Dollars per Gallon":"Price per Gallon"})
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df['Year'] = cali_all_form_retail_gas_prices_df['Year-Month'].dt.year
cali_all_form_retail_gas_prices_df['Month'] = cali_all_form_retail_gas_prices_df['Year-Month'].dt.month
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df.columns
```

```python
cali_all_form_retail_gas_prices_df= cali_all_form_retail_gas_prices_df.drop(["Year-Month"],axis=1)
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df= pd.DataFrame(data =cali_all_form_retail_gas_prices_df,
                                        columns={'Year':'Year',
                                                 'Month':'Month',
                                                'Price per Gallon': 'Price per Gallon'})
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df = cali_all_form_retail_gas_prices_df[(cali_all_form_retail_gas_prices_df.Year >= int(2001))]
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df = cali_all_form_retail_gas_prices_df[2:]
cali_all_form_retail_gas_prices_df
# cali_all_form_retail_gas_prices_df = cali_all_form_retail_gas_prices_df.reset_index()
# cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df = cali_all_form_retail_gas_prices_df.reset_index()
cali_all_form_retail_gas_prices_df
```

```python
cali_all_form_retail_gas_prices_df = cali_all_form_retail_gas_prices_df.drop(columns=["level_0", "index"])
cali_all_form_retail_gas_prices_df
```

```python
import os
cali_all_form_retail_gas_prices_df.to_csv('Resources/cali_all_form_retail_gas_prices.csv', index=False)
```

# Work on Price of U.S. Natural Gas Imports

```python
imports_price_df = pd.read_csv('../Resources/Datasets/Price_of_U.S._Natural_Gas_Imports.csv',skiprows=4)
imports_price_df.head(10)
```

# Merge Exports_df and imports_df

```python
us_price_export_import = imports_price_df.merge(exports_price_df, how='left', on = 'Month')
us_price_export_import
```

```python
us_price_export_import["Date"] = pd.to_datetime(us_price_export_import["Month"])
us_price_export_import.head()
```

```python
us_price_export_import= us_price_export_import.drop(["Month"],axis=1)
us_price_export_import
```

```python
us_price_export_import.columns
```

```python
us_price_export_import = pd.DataFrame(data=us_price_export_import, columns=['Date',
            "Price of U.S. Natural Gas Imports  Dollars per Thousand Cubic Feet",
            "Price of U.S. Natural Gas Exports  Dollars per Thousand Cubic Feet"])
us_price_export_import
```

```python
us_price_export_import=us_price_export_import.rename(columns={"Date":"Year-Month",
        "Price of U.S. Natural Gas Imports  Dollars per Thousand Cubic Feet":"Imports_Price_per_Gallon",
        "Price of U.S. Natural Gas Exports  Dollars per Thousand Cubic Feet":"Exports_Price_per_Gallon"})
us_price_export_import
```

```python
us_price_export_import['Year'] = us_price_export_import['Year-Month'].dt.year
us_price_export_import['Month'] = us_price_export_import['Year-Month'].dt.month
us_price_export_import
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# gca stands for 'get current axis'
ax = plt.gca()

us_price_export_import.plot(kind='line',x='Year-Month',y='Imports_Price_per_Gallon',ax=ax)
us_price_export_import.plot(kind='line',x='Year-Month',y='Exports_Price_per_Gallon', color='red', ax=ax)

plt.show()
```

```python
us_price_export_import.columns
```

```python
us_price_export_import= us_price_export_import.drop(["Year-Month"],axis=1)
us_price_export_import
```

```python
us_price_export_import= pd.DataFrame(data =us_price_export_import,
                                        columns={'Year':'Year',
                                                 'Month':'Month',
                                                'Exports_Price_per_Gallon': 'Exports_Price_per_Gallon',
                                                'Imports_Price_per_Gallon': 'Imports_Price_per_Gallon'})
us_price_export_import
```
